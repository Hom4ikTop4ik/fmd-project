```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_correlation_matrix(df):
    corr = df.corr()
    plt.figure(figsize=(20,20))
    sns.heatmap(corr, annot=False, cmap="coolwarm", linewidths=.01)
    plt.show()

plot_correlation_matrix(df)
```
